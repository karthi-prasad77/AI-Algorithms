## Normal Distribution Formula

The probability density function (PDF) of a normal distribution is:

\[
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
\]

Where:  
- **\( x \)** = Value in the distribution  
- **\( \mu \) (mean)** = Center of the distribution  
- **\( \sigma \) (standard deviation)** = Spread of the data  
- **\( \pi \)** = Pi (≈ 3.14159)  
- **\( e \)** = Euler's number (≈ 2.718)  


```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parameters for normal distribution
mean = 0 # center of the distribution
std = 1 # spread of the distribution
num_samples = 1000 # number of data points
```

```python
# generate data
data = np.random.normal(mean, std, num_samples)
```

```python
# plot histogram to visualize distribution
plt.hist(data, bins=30, density=True, alpha=0.6, color='b')
```

```python
# plot the actual normal distribution curve
x = np.linspace(-4, 4, 10000)

x
```

```python
pdf = (1 / std * (np.sqrt(2 * np.pi))) * np.exp(-((x - mean) ** 2) / (2 * std ** 2))

pdf
```

```python
plt.plot(x, pdf, 'r', linewidth=2)
```

```python
# Plot histogram to visualize distribution
plt.hist(data, bins=30, density=True, alpha=0.6, color='b')

# Plot the actual normal distribution curve
x = np.linspace(-4, 4, 1000)
pdf = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-((x - mean) ** 2) / (2 * std ** 2))
plt.plot(x, pdf, 'r', linewidth=2)

# Labels and title
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Normal Distribution")

# Show plot
plt.show()
```
